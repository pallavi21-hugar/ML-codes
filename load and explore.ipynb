```python
import pandas as pd
    
#define the file paths
csv_file_path="C:\\Users\GodMode\Documents\\sample1.csv"
excel_file_path="C:\\Users\GodMode\Documents\\sample.xlsx"
#load the csv file
data_csv=pd.read_csv(csv_file_path)
print("csv file data:")
print(data_csv)

#load the excel file
data_excel=pd.read_excel(excel_file_path)
print("\n excel file data:")
print(data_excel)

#basic data exploration
print("\n data description:")
print("csv data description:")
print(data_csv.describe())
print("\n excel data description:")
print(data_excel.describe())

#displaying data types
print("\n data types in excel file:")
print(data_excel.dtypes)
print("\n data types in csv file:")
print(data_csv.dtypes)

print("dataset information:")
print(data_csv.info())

print("\n first few rows of the dataset:")
print(data_csv.head())
print("\n first few rows of the dataset:")
print(data_excel.head())

print("\n last few rows of the dataset:")
print(data_csv.tail())
print("\n last few rows of the dataset:")
print(data_excel.tail())





```
